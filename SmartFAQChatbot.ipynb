```python

!pip -q install transformers sentencepiece torch


from transformers import AutoTokenizer, AutoModelForSeq2SeqLM


model_name = "google/flan-t5-base"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)


def ask(question):
    q = question.lower()


    if "machine learning" in q:
        return "Machine learning is a branch of artificial intelligence that allows computers to learn from data and make predictions."

    elif "deep learning" in q:
        return "Deep learning is a subset of machine learning that uses neural networks with multiple layers."

    elif "data analytics" in q or "data analysis" in q:
        return "Data analytics is the process of analyzing data to find patterns and useful insights."

    elif "computer" in q:
        return "A computer is an electronic device that processes data and performs tasks using programs."


    prompt = f"""
You are a helpful academic assistant.

Answer clearly in 2 simple sentences.
Do NOT generate random or meaningless text.
If unsure, give a general correct definition.

Question: {question}
Answer:
"""

    inputs = tokenizer(prompt, return_tensors="pt", truncation=True)

    outputs = model.generate(
        inputs["input_ids"],
        max_length=60,
        do_sample=False,
        num_beams=5
    )

    answer = tokenizer.decode(outputs[0], skip_special_tokens=True)


    if len(answer.split()) < 5 or "http" in answer.lower():
        return "This is a topic related to computer science and artificial intelligence. It involves concepts used in academic study."

    return answer



print("🎓 Smart FAQ Chatbot (FREE - No API)\n")

while True:
    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("Chatbot: Goodbye 👋")
        break

    response = ask(user_input)
    print("Chatbot:", response)
```
